```python
import matplotlib.pyplot as plt

def load_data(file_name):
    sizes = []
    series1 = []
    series2 = []

    with open(file_name, 'r') as file:
        for line in file:
            if line.strip():
                values = line.split()
                sizes.append(int(values[0]))
                series1.append(int(values[1]))
                series2.append(int(values[2]))

    return sizes, series1, series2

file_name = "performance.txt"
sizes, series1, series2 = load_data(file_name)

plt.figure(figsize=(10, 6))
plt.scatter(sizes, series1, label="sequential", marker='o', color = 'g', linestyle="--")
plt.scatter(sizes, series2, label="parallel", marker='o', color = 'r', linestyle="--")

plt.xscale('log')
plt.xlabel('Size')
plt.ylabel('Time (ms)')
plt.legend()

plt.savefig('plot.png', format='png')
plt.show()
```
