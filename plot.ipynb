```python
import matplotlib.pyplot as plt
import numpy as np

x = [32, 64, 128, 256, 512, 1024, 2048, 4096, 8192]
y = [0.65, 1.61, 4.34, 14.69, 57.37, 229.91, 928.94, 3799.96, 15220.38]

plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o')
plt.xscale('log')
plt.yscale('log')

plt.title("Time in function of size", fontsize=14)
plt.xlabel("Size (log)", fontsize=12)
plt.ylabel("Time (log)", fontsize=12)
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.savefig('plot.png')
```
